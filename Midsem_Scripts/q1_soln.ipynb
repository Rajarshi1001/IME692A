```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("q1.csv")
data.head(10)
```

```python
def regline(X, b1, b0):
    y = b1*X + b0
    plt.figure(figsize = (10,7))
    plt.xlabel("GRE")
    plt.ylabel("CGPA")
    plt.title("CGPA vs GRE")
    plt.scatter(GRE ,CGPA)
    plt.plot(X, y, color = "red")
    plt.grid()
    plt.show()
    pearson_corr = np.corrcoef(GRE, CGPA)[0, 1]
    print(f"Pearson correlation coefficient: {pearson_corr}")
    return y

GRE = data["GRE"].to_numpy()
CGPA = data["CGPA"].to_numpy()
# Calculate means of GRE and CGPA
mean_GRE = sum(GRE) / len(GRE)
mean_CGPA = sum(CGPA) / len(CGPA)

# Calculate b1 (slope)
numerator = sum([(GRE[i] - mean_GRE) * (CGPA[i] - mean_CGPA) for i in range(len(GRE))])
denominator = sum([(GRE[i] - mean_GRE) ** 2 for i in range(len(GRE))])
b1 = numerator / denominator

# Calculate b0 (y-intercept)
b0 = mean_CGPA - b1 * mean_GRE

print(f"b0 (y-intercept) = {b0}")
print(f"b1 (slope) = {b1}")

res = regline(GRE, b1, b0)
```
